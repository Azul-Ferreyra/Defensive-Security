Ataques estadisticos: Analisis de frecuencia 

```python
def ngramas(texto,n):
    texto = texto.replace(" ","")
    return[texto[i:i+n] for i in range(len(texto)-n+1)]
```

```python
ngramas("esto es un texto de prueba", 2)
```

```python
from collections import Counter
import pandas as pd 
```

```python
def frec_ngramas(texto,n):
    ng = ngramas(texto,n)
    frecuencias = Counter(ng)
    df = pd.DataFrame(frecuencias.most_common(), columns= ["letra","frecuencias"])
    return df
```

```python
frec_ngramas("este es un texto de prueba para determinal la frecuencia en la que aparecen los ngramas", 2)
```

```python
def leer_frecuencias(archivo):
    return pd.read_csv(archivo, sep= " " , names= ["letras", "frecuencias"])
```

```python
leer_frecuencias("spanish_monograms.txt")
```

```python
def leer_texto_cifrado(texto_cifrado, n):
    with open (texto_cifrado) as f:
        texto_cifrado = f.readlines()[0]
        return frec_ngramas(texto_cifrado, n )
```

```python
alfabeto_permutado = {
 'a': 'u',
 'b': 'q',
 'c': 't',
 'd': 'f',
 'e': 'v',
 'f': 'n',
 'g': 'r',
 'h': 'a',
 'i': 'x',
 'j': 'b',
 'k': 'd',
 'l': 'e',
 'm': 'y',
 'n': 'j',
 'o': 'p',
 'p': 'c',
 'q': 'h',
 'r': 'z',
 's': None,
 't': 'i',
 'u': 'l',
 'v': 's',
 'w': 'g',
 'x': 'o',
 'y': None,
 'z': 'm'
}
```

```python
def algoritmo_descifrado(texto_cifrado, alfabeto_permutado):
    texto_plano = ""
    with open (texto_cifrado) as f:
        texto_cifrado = f.readlines()[0]

    for letra in texto_cifrado:
        if not alfabeto_permutado[letra.lower()]:
            texto_plano += letra
        else:
            texto_plano += alfabeto_permutado[letra.lower()]
    return texto_plano
```

```python
algoritmo_descifrado('texto_cifrado.txt', alfabeto_permutado)
```
